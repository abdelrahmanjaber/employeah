```python
import re
import pandas as pd

MAX_SKILLS = 20

# Load known skills
known_skills = [
    line.strip()
    for line in open("skills.txt", "r", encoding="utf-8").readlines()
    if line.strip()
]

# Build regex
sorted_skills = sorted(known_skills, key=len, reverse=True)
regex = r"\b(?:" + "|".join(re.escape(s) for s in sorted_skills) + r")\b"
pattern = re.compile(regex, re.IGNORECASE)

# Load dataset
df = pd.read_csv("/Users/alessandromolinarroet/Desktop/web_application/Scapers/arbeitnow_jobs.csv")

# Extract skills and append to dataset
skill_columns = [f"skill{i}" for i in range(1, MAX_SKILLS + 1)]
for col in skill_columns:
    df[col] = ""

for idx, description in enumerate(df["description"]):
    matches = pattern.findall(str(description))
    print(matches)
    extracted = sorted({m.title() for m in matches})[:MAX_SKILLS]

    for i, skill in enumerate(extracted):
        df.at[idx, f"skill{i+1}"] = skill
```

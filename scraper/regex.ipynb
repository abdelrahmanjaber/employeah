```python
import ast

def to_list(x):
    if isinstance(x, list):
        return x
    if x is None or x != x: 
        return []
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return [i.strip() for i in x.strip("[]").split(",")]
    return [x]

df["tags"] = df["tags"].apply(to_list)


skill_cols = [f"skill{i}" for i in range(1, 21)]
tag_dict = {}

for _, row in df.iterrows():
    tags = row["tags"]

    skills = [row[col] for col in skill_cols if col in df.columns and row[col] and str(row[col]).strip()]

    for t in tags:
        if t not in tag_dict:
            tag_dict[t] = {}  
        for skill in skills:
            if skill in tag_dict[t]:
                tag_dict[t][skill] += 1
            else:
                tag_dict[t][skill] = 1
```
